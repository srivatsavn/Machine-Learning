```python
import pandas as pd
import numpy as np
movies = pd.read_csv('C:/Users/Administrator/Desktop/Datasets/movies.csv')
ratings = pd.read_csv('C:/Users/Administrator/Desktop/Datasets/ratings.csv')
ratings.head()
```

```python
ratings['userId'] = 'User-' + ratings['userId'].astype(str)
ratings['movieId'] = 'Movie-' + ratings['movieId'].astype(str)
ratings.head(5)
```

```python
##Create user-movie table
print('No of Unique Movies %d' %ratings['movieId'].nunique())
print('No of Unique users %d' %ratings['userId'].nunique())
```

```python
user_item_matrix = ratings.pivot_table(index='userId',
                                      columns='movieId',
                                      values='rating')
print(user_item_matrix.shape)
user_item_matrix.head()
```

```python
# 1 Average number of ratings per user
print(user_item_matrix.count(axis=1).mean())

#2 User with most number of ratings(jobless user)
print(user_item_matrix.count(axis=1).sort_values(ascending=False).head(1))

#Average number of ratings per movie
print(user_item_matrix.count().mean())

#Movies with most number of ratings (most discussed movie)
print(user_item_matrix.count().sort_values(ascending=False).head(1))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

#cosine_similarity([[1,2,3],[5,6,7]])

```

```python
##Find out row mean and subtract with each cell and impute zeros


ui_matrix = user_item_matrix.apply(lambda v: v - v.mean(),axis=1).fillna(0)

ui_matrix.head()
```

```python
sim_matrix= pd.DataFrame(cosine_similarity(ui_matrix),index=ui_matrix.index,
            columns = ui_matrix.index)
sim_matrix.head()
```

```python
neighs_user100 = sim_matrix['User-100'].drop('User-100').sort_values(ascending=False).head(7).index
neighs_user100
```

```python
user_item_matrix.loc['User-100']
```

```python
preds = pd.DataFrame(columns=['mid','predicted_rating'])
for mid,rating in user_item_matrix.loc['User-100'].items():
    if pd.isnull(rating):
        #print(mid)
        mid_ratings = user_item_matrix.loc[neighs_user100,mid]
        avg_mid_ratings = mid_ratings.mean()
        preds = preds.append({'mid':mid,
                             'predicted_rating':avg_mid_ratings},
                            ignore_index=True)
        
        #print(mid,mid_ratings.values,avg_mid_ratings)
preds.sort_values('predicted_rating',ascending=False).head(3)        
```

```python
def ubcf(userid):
    preds = pd.DataFrame(columns=['mid','predicted_rating'])
    neighs_user = sim_matrix[userid].drop(userid).sort_values(ascending=False).head(7).index
    for mid,rating in user_item_matrix.loc['User-100'].items():
        if pd.isnull(rating):
            mid_ratings = user_item_matrix.loc[neighs_user100,mid]
            avg_mid_ratings = mid_ratings.mean()
            preds = preds.append({'mid':mid,
                             'predicted_rating':avg_mid_ratings},
                            ignore_index=True)
    preds['movieid'] = preds['mid'].str.replace('Movie-','').astype(int)

    preds = pd.merge(preds,movies,
        left_on='movieid',
        right_on='movieId')

    movies_recom = preds.sort_values('predicted_rating',ascending=False).head(3)['title']
    return movies_recom.tolist()
        
```
