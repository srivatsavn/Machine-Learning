```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

whole = pd.read_csv('C:/Users/Administrator/Desktop/Datasets/Wholesale customers data.csv')

whole = whole.drop(['Channel','Region'],axis = 1)

kmean_whole = KMeans(n_clusters = 3,random_state=100)
kmean_whole.fit(whole)
label_whole = kmean_whole.predict(whole)
label_whole

whole['Labels'] = label_whole
whole

centroids = kmean_whole.cluster_centers_
```

```python
from sklearn.metrics import silhouette_score
```

```python
sc = silhouette_score(whole, label_whole)
```

```python
sc
```

```python
k=[2,3,4,5,6,8]
for i in k:
    kmean_whole = KMeans(n_clusters = i, random_state=100)
    kmean_whole.fit(whole)
    label_whole = kmean_whole.predict(whole)
    whole['Labels'] = label_whole
    sc = silhouette_score(whole, label_whole)
    print(sc)
```

```python
print("This is SC:", sc)
```

```python
kmean_whole.inertia_
```

```python
k=[1,2,3,4,5,6,7,8]
for i in k:
    kmean_whole = KMeans(n_clusters = i, random_state=100)
    kmean_whole.fit(whole)
    label_whole = kmean_whole.predict(whole)
    whole['Labels'] = label_whole
    iner = kmean_whole.inertia_
    print(iner)
```
