```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
mush = pd.read_csv('C:/Users/Administrator/Desktop/Datasets/mushroom.csv')
```

```python
mush.head()
```

```python
mush_new = mush.drop('class', axis =1)
```

```python
mush_dummies = pd.get_dummies(mush_new)
```

```python
mush_dummies['class'] = mush['class']
```

```python
mush_dummies.shape
```

```python
train, test = train_test_split(mush_dummies, test_size = 0.3, random_state = 100)
```

```python
train_y = train['class']
test_y = test['class']
train_x = train.drop('class', axis = 1)
test_x = test.drop('class', axis = 1)

model = DecisionTreeClassifier(random_state=100, max_depth=3)
model.fit(train_x,train_y)
```

```python
test_pred = model.predict(test_x)
print(len(test_pred))
test_pred[:5]
```

```python
df_pred = pd.DataFrame({'actual' : test_y,
                       'predicted': test_pred})
df_pred.head()
```

```python
df_pred['pred_status'] = df_pred['actual'] == df_pred['predicted']
df_pred.head()
```

```python
df_pred['pred_status'].value_counts()
```

```python
df_pred['pred_status'].value_counts() / df_pred.shape[0] * 100
```

# RANDOM FOREST

```python
all_cols = train.drop('class', axis=1).columns
cols_used = all_cols[np.random.randint(1, len(all_cols), 3)]
print(cols_used)

train_y = train['class']
test_y = test['class']

train_x = train[cols_used]
test_x = test[cols_used]

model = DecisionTreeClassifier(random_state=100)
model.fit(train_x,train_y)
pd.DataFrame(model.predict(test_x), index=test.index, columns=['Attrition_Predicted'])
```

```python
train_y = train['class']
test_y = test['class']

train_x = train.drop('class', axis = 1)
test_x = test.drop('class', axis = 1)

from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(random_state=100)
model_rf.fit(train_x, train_y)

test_pred = model_rf.predict(test_x)
df_pred = pd.DataFrame({'actual' : test_y,
                       'predicted' : test_pred})
df_pred['pred_status'] = df_pred['actual'] == df_pred['predicted']
df_pred['pred_status'].sum() / df_pred.shape[0] * 100
```
