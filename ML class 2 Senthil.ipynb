# Regularization

## https://towardsdatascience.com/regularization-in-machine-learning-76441ddcf99a

```python
import pandas as pd
import numpy as np
market = pd.read_csv("C:/Users/Administrator/Desktop/Datasets/market_data.csv")
data = market.drop(['Item_Identifier','Outlet_Identifier'], axis=1)
data
data.columns

data['Outlet_Establishment_Year'] = 2010 - data['Outlet_Establishment_Year']
data

data.isna().sum()

data['Item_Weight'] = data['Item_Weight'].fillna(method = 'bfill')
data['Outlet_Size'] = data['Outlet_Size'].fillna(method = 'ffill')

from sklearn.preprocessing import LabelEncoder

imp = LabelEncoder()
data['Item_Fat_Content'] = imp.fit_transform(data['Item_Fat_Content'])
data['Outlet_Size']=imp.fit_transform(data['Outlet_Size'])
data['Outlet_Location_Type']=imp.fit_transform(data['Outlet_Location_Type'])

data1 = pd.get_dummies(data)
data1

target=data1['Item_Outlet_Sales']
data1.drop(['Item_Outlet_Sales'],axis=1,inplace=True)
 

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

[xtrain, xval, ytrain, yval] = train_test_split(data1, target, test_size=0.3, random_state=0)

regmod = LinearRegression()
regmod.fit(xtrain, ytrain)
ypred = regmod.predict(xval)

mse=mean_squared_error(yval, ypred)
mse
rmse=np.sqrt(mse)
rmse
rsq=regmod.score(xval,yval)
rsq

coeff=regmod.coef_
coeff

min(data['Item_Outlet_Sales'])
max(data['Item_Outlet_Sales'])

```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
x_plot = plt.scatter(ypred, (ypred - yval), c='b')
plt.hlines(y=0, xmin= -1000, xmax=5000)
plt.title('Residual Plot')
predictors = xtrain.columns 
```

```python
coef = pd.Series(regmod.coef_, predictors).sort_values()
plt.figure(figsize=(10,8))
plt.figure(1)
coef.plot(kind='bar', title='Model Coefficeient')
plt.tight_layout()
```

```python
from sklearn.linear_model import Ridge
```

```python
##training the model

ridgeReg = Ridge(alpha=0.05, normalize=True)
ridgeReg.fit(xtrain,ytrain)
ridge_pred = ridgeReg.predict(xval)
mse_rid = np.sqrt(np.mean((ridge_pred-yval)**2))
rsq_ridge = ridgeReg.score(xval,yval)

coef_rig = pd.Series(ridgeReg.coef_,predictors).sort_values()
plt.figure(2)
plt.figure(figsize=(10,8))
coef_rig.plot(kind='bar', title='Model Coefficient')
```

```python
## Alpha value as 0

ridgeReg = Ridge(alpha=0, normalize=True)
ridgeReg.fit(xtrain,ytrain)
ridge_pred = ridgeReg.predict(xval)
mse_rid = np.sqrt(np.mean((ridge_pred-yval)**2))
rsq_ridge = ridgeReg.score(xval,yval)

coef_rig = pd.Series(ridgeReg.coef_,predictors).sort_values()
plt.figure(2)
plt.figure(figsize=(10,8))
coef_rig.plot(kind='bar', title='Model Coefficient')
```
